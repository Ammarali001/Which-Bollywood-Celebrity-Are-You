```python
import os
import random
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
DATASET_PATHS = [
    "/content/Bollywood_celeb_face_localized/bollywood_celeb_faces2",
    "/content/Bollywood_celeb_face_localized/bollywood_celeb_faces_0",
    "/content/Bollywood_celeb_face_localized/bollywood_celeb_faces_1"
]
```

```python
for path in DATASET_PATHS:
    print(path, "✅ Exists" if os.path.exists(path) else "❌ Not Found")
```

```python
all_classes = set()

for dataset in DATASET_PATHS:
    for folder in os.listdir(dataset):
        folder_path = os.path.join(dataset, folder)

        if os.path.isdir(folder_path):
            all_classes.add(folder)

print(f"Total Unique Celebrities: {len(all_classes)}")
```

```python
image_counts = {}

for dataset in DATASET_PATHS:
    for celeb in os.listdir(dataset):

        celeb_path = os.path.join(dataset, celeb)

        if not os.path.isdir(celeb_path):
            continue

        images = [
            img for img in os.listdir(celeb_path)
            if img.lower().endswith((".jpg", ".jpeg", ".png"))
        ]

        image_counts[celeb] = image_counts.get(celeb, 0) + len(images)

print(f"Total Celebrities : {len(image_counts)}")
print(f"Total Images      : {sum(image_counts.values())}")
```

```python
df = pd.DataFrame({
    "Celebrity": image_counts.keys(),
    "Images": image_counts.values()
})

df = df.sort_values("Images", ascending=False)

df.head(10)
```

```python
print("Dataset Statistics")
print("-" * 40)

print("Total Celebrities :", len(df))
print("Total Images      :", df["Images"].sum())
print("Average Images    :", round(df["Images"].mean(), 2))
print("Minimum Images    :", df["Images"].min())
print("Maximum Images    :", df["Images"].max())
print("Median Images     :", df["Images"].median())
```

```python
plt.figure(figsize=(18,6))

plt.bar(df["Celebrity"], df["Images"])

plt.xticks(rotation=90)

plt.xlabel("Celebrity")
plt.ylabel("Number of Images")
plt.title("Images per Celebrity")

plt.show()
```

```python
fig, axes = plt.subplots(3, 5, figsize=(15, 10))

for ax in axes.ravel():

    while True:

        dataset = random.choice(DATASET_PATHS)
        celeb = random.choice(os.listdir(dataset))

        celeb_path = os.path.join(dataset, celeb)

        if not os.path.isdir(celeb_path):
            continue

        images = [
            img for img in os.listdir(celeb_path)
            if img.lower().endswith((".jpg", ".jpeg", ".png"))
        ]

        if len(images) == 0:
            continue

        img_path = os.path.join(celeb_path, random.choice(images))

        break

    img = Image.open(img_path)

    ax.imshow(img)
    ax.set_title(celeb, fontsize=9)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
data = []

for dataset in DATASET_PATHS:

    for celebrity in os.listdir(dataset):

        celeb_path = os.path.join(dataset, celebrity)

        if not os.path.isdir(celeb_path):
            continue

        for image in os.listdir(celeb_path):

            if image.lower().endswith((".jpg", ".jpeg", ".png")):

                data.append({
                    "filepath": os.path.join(celeb_path, image),
                    "label": celebrity
                })

df = pd.DataFrame(data)

print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
print("Before:", len(df))

df = df.drop_duplicates(subset="filepath")

print("After :", len(df))
```

```python
from PIL import Image

valid_images = []

for path in df["filepath"]:

    try:
        img = Image.open(path)
        img.verify()
        valid_images.append(True)

    except:
        valid_images.append(False)

df = df[valid_images].reset_index(drop=True)

print("Remaining Images:", len(df))
```

```python
sizes = []

for path in df["filepath"][:1000]:

    img = Image.open(path)
    sizes.append(img.size)

sizes[:10]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["label_encoded"] = encoder.fit_transform(df["label"])

df.head()
```

```python
import pickle

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)

print("Saved Successfully")
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)

print("Training :", len(train_df))
print("Validation:", len(val_df))
print("Testing :", len(test_df))
```

```python
print(f"Train : {len(train_df)/len(df):.2%}")
print(f"Validation : {len(val_df)/len(df):.2%}")
print(f"Test : {len(test_df)/len(df):.2%}")
```

```python
import tensorflow as tf

print("TensorFlow Version:", tf.__version__)
```

```python
# Image Size
IMG_SIZE = (224, 224)

# Batch Size
BATCH_SIZE = 32

# Number of Classes
NUM_CLASSES = len(encoder.classes_)

print("Classes:", NUM_CLASSES)
```

```python
def load_image(filepath, label):

    image = tf.io.read_file(filepath)

    image = tf.image.decode_jpeg(image, channels=3)

    image = tf.image.resize(image, IMG_SIZE)

    image = image / 255.0

    return image, label
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.15),
    tf.keras.layers.RandomZoom(0.15),
    tf.keras.layers.RandomContrast(0.10),
])
```

```python
train_ds = tf.data.Dataset.from_tensor_slices(
    (
        train_df["filepath"].values,
        train_df["label_encoded"].values
    )
)

train_ds = train_ds.map(load_image,
                        num_parallel_calls=tf.data.AUTOTUNE)
```

```python
train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y),
    num_parallel_calls=tf.data.AUTOTUNE
)
```

```python
val_ds = tf.data.Dataset.from_tensor_slices(
    (
        val_df["filepath"].values,
        val_df["label_encoded"].values
    )
)

val_ds = (
    val_ds
    .map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
test_ds = tf.data.Dataset.from_tensor_slices(
    (
        test_df["filepath"].values,
        test_df["label_encoded"].values
    )
)

test_ds = (
    test_ds
    .map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)
```
